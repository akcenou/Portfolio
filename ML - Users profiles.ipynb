````python
# –¥–∞–Ω–Ω—ã–µ –∑–∞ –º–∞–π
df_may = (df
          .query('year_month == "2019-05"')       
          .groupby('category')['order_id'].nunique().reset_index().sort_values(by='category', ascending=False)
         )
# –¥–∞–Ω–Ω—ã–µ –±–µ–∑ –º–∞—è
df_without_may = (df.query('year_month != "2019-05"') 
                    .groupby('category')['order_id'].nunique().reset_index().sort_values(by='category', ascending=False)
                 )
df_without_may['order_id'] = df_without_may['order_id'] /(months - 1)

# —Å–≤–æ–¥–Ω—ã–π –¥–∞—Ç–∞—Å–µ—Ç
df_compare = pd.merge(df_without_may, df_may, on='category', how='inner')
df_compare.columns = ['category', 'avg_orders', 'orders May 2019']
df_compare['diff'] = (df_compare['orders May 2019'] / df_compare['avg_orders'] - 1) * 100
df_compare
````

````python
# –∫–æ–ª-–≤–æ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –∏ —Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª-–≤–æ –∑–∞–∫–∞–∑–æ–≤ –≤ –º–µ—Å—è—Ü
pd.set_option('display.float_format', '{:.4f}'.format)

customers_grouped = df.groupby('customer_id')['order_id'].nunique().reset_index()
customers_grouped['avg_purchases_month'] = customers_grouped['order_id'] / months

customers_grouped.rename(columns={'order_id':'total_orders'}, inplace=True)
customers_grouped
````

–ö–ª–∞—Å—Ç–µ—Ä 2 - '*–†–µ–¥–∫–∏–π –∫—Ä—É–ø–Ω—ã–π –∫–ª–∏–µ–Ω—Ç*': —Ä–µ–¥–∫–æ –ø—Ä–æ–∏–∑–≤–æ–¥—è—Ç –±–æ–ª–µ–µ 1 –∑–∞–∫–∞–∑–∞, –Ω–æ –ø—Ä–∏–æ–±—Ä–µ—Ç–∞—é—Ç –¥–æ—Ä–æ–≥–æ–π —Ç–æ–≤–∞—Ä —Å —Ü–µ–Ω–æ–π 2773 —É.–µ, –æ–±—â–∏–π —á–µ–∫ 3221 —É.–µ., –≤ –∑–∞–∫–∞–∑–µ –ø–æ—á—Ç–∏ –≤—Å–µ–≥–¥–∞ –±–æ–ª–µ–µ 1 –µ–¥. —Ç–æ–≤–∞—Ä–∞

````python
# –ø–µ—Ä–µ–∏–º–µ–Ω—É–µ–º –∫–ª–∞—Å—Ç–µ—Ä—ã –∏ –¥–æ–±–∞–≤–∏–º –∏—Ö –≤ –æ–±—â–∏–π –¥–∞—Ç–∞—Ñ—Ä–µ–π–º
pd.set_option('display.float_format', None)

df_clustered = pd.merge(df, customers[['customer_id','cluster_km']], on='customer_id',how='left')
df_clustered['cluster_km'] = df_clustered['cluster_km'].replace({0:'–ß–∞—Å—Ç—ã–π –º–µ–ª–∫–∏–π –∫–ª–∏–µ–Ω—Ç', 1:'–°–∫–ª–æ–Ω–Ω—ã–π –∫ –æ—Ç—Ç–æ–∫—É', 2:'–†–µ–¥–∫–∏–π –∫—Ä—É–ø–Ω—ã–π –∫–ª–∏–µ–Ω—Ç'})
df_clustered.info()
````

#### –û–ø–∏—Å–∞–Ω–∏–µ —Å–µ–≥–º–µ–Ω—Ç–æ–≤

–í–∑–≥–ª—è–Ω–µ–º, —á—Ç–æ –ø–æ–∫—É–ø–∞—é—Ç —Ä–∞–∑–¥–µ–ª–µ–Ω–Ω—ã–µ –Ω–∞ –∫–ª–∞—Å—Ç–µ—Ä—ã –ø–æ–∫—É–ø–∞—Ç–µ–ª–∏:

````python
# –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–∫–∞–∑–æ–≤ –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∫–ª–∞—Å—Ç–µ—Ä–∞
pivot_orders = df_clustered.pivot_table(index='category', columns='cluster_km', values='order_id', aggfunc='nunique')

colors = plt.cm.Dark2(range(len(pivot_orders.columns)))
pivot_orders.plot(kind='bar', figsize=(15, 6), color=colors)

plt.title('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–∫–∞–∑–æ–≤ –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∫–ª–∞—Å—Ç–µ—Ä–∞')
plt.xlabel('–ö–∞—Ç–µ–≥–æ—Ä–∏—è —Ç–æ–≤–∞—Ä–∞')
plt.xticks(rotation=0, fontsize=12)
plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–∫–∞–∑–æ–≤')
plt.legend(title='–ö–ª–∞—Å—Ç–µ—Ä')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()  
plt.show()
````

- *–†–µ–¥–∫–∏–π –∫—Ä—É–ø–Ω—ã–π –∫–ª–∏–µ–Ω—Ç*: –ø–æ–∫—É–ø–∞–µ—Ç, –∫–∞–∫ –ø—Ä–∞–≤–∏–ª–æ, –±–æ–ª–µ–µ –¥–æ—Ä–æ–≥–∏–µ —Ç–æ–≤–∞—Ä—ã "–¥–ª—è –¥–æ–º–∞" –∏ "—Ç–µ–∫—Å—Ç–∏–ª—å"

- *–°–∫–ª–æ–Ω–Ω—ã–π –∫ –æ—Ç—Ç–æ–∫—É*: –µ–¥–∏–Ω–∏—á–Ω—ã–µ –Ω–∏–∑–∫–æ–¥–æ—Ö–æ–¥–Ω—ã–µ –ø–æ–∫—É–ø–∫–∏, –≤ –±–æ–ª—å—à–∏–Ω—Å—Ç–≤–µ —Å–≤—è–∑–∞–Ω–Ω—ã–µ —Å "—Ü–≤–µ—Ç–∞–º–∏", "—Ä–∞—Å—Å–∞–¥–æ–π", "–¥–µ–∫–æ—Ä–æ–º", –∞ —Ç–∞–∫–∂–µ —Ç–æ–≤–∞—Ä–∞–º–∏ "–¥–ª—è –¥–æ–º–∞" –∏ "—Ç–µ–∫—Å—Ç–∏–ª–µ–º"

–û—Å—Ç–∞–≤—à–∏–µ—Å—è –∫–ª–∞—Å—Ç–µ—Ä—ã –ø–æ–∫–∞–∑—ã–≤–∞—é—Ç –Ω–µ—Å—É—â–µ—Å—Ç–≤–µ–Ω–Ω—É—é —Ä–∞–∑–Ω–∏—Ü—É –≤ —Å—Ä–µ–¥–Ω–µ–º —á–µ–∫–µ –º–µ–∂–¥—É —Å–æ–±–æ–π

````python
# –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–∫–∞–∑–æ–≤ –ø–æ –º–µ—Å—è—Ü–∞–º –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∫–ª–∞—Å—Ç–µ—Ä–∞
pivot_season = df_clustered.pivot_table(index='cluster_km', columns='year_month', values='order_id', aggfunc='nunique')

pivot_season.plot(kind='bar', figsize=(15, 5))

plt.title('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–∫–∞–∑–æ–≤ –ø–æ –º–µ—Å—è—Ü–∞–º –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∫–ª–∞—Å—Ç–µ—Ä–∞')
plt.xlabel('–ö–ª–∞—Å—Ç–µ—Ä')
plt.xticks(rotation=0, fontsize=12)
plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–∫–∞–∑–æ–≤')
plt.legend(title='–ö–ª–∞—Å—Ç–µ—Ä')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()  
plt.show()
````

- *–ß–∞—Å—Ç—ã–π –º–µ–ª–∫–∏–π –∫–ª–∏–µ–Ω—Ç*: –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç –ø–æ–∫—É–ø–∫–∏ –≤ —Ç–µ—á–µ–Ω–∏–µ –≤—Å–µ–≥–æ –ø–µ—Ä–∏–æ–¥–∞ —Å –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –Ω–µ–±–æ–ª—å—à–∏–º–∏ –ø–æ–¥—ä–µ–º–∞–º–∏/–ø–∞–¥–µ–Ω–∏—è–º–∏.

- *–†–µ–¥–∫–∏–π –∫—Ä—É–ø–Ω—ã–π –∫–ª–∏–µ–Ω—Ç*: –ø–æ–∫—É–ø–∞–µ—Ç –≤ —Ç–µ—á–µ–Ω–∏–µ –≤—Å–µ–≥–æ –ø–µ—Ä–∏–æ–¥–∞ —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–µ–µ –≤—Å–µ–≥–æ, –±–µ–∑ –≤—ã—Ä–∞–∂–µ–Ω–Ω—ã—Ö —Å–∫–∞—á–∫–æ–≤

````python
clusters = ['–ß–∞—Å—Ç—ã–π –º–µ–ª–∫–∏–π –∫–ª–∏–µ–Ω—Ç', '–†–µ–¥–∫–∏–π –∫—Ä—É–ø–Ω—ã–π –∫–ª–∏–µ–Ω—Ç', '–°–∫–ª–æ–Ω–Ω—ã–π –∫ –æ—Ç—Ç–æ–∫—É']

for cl in clusters:
    pivot_category = (df_clustered[df_clustered['cluster_km'] == cl]
                      .pivot_table(index='year_month', columns='category', values='order_id', aggfunc='nunique')
                      )
    pivot_category.plot(kind='bar', figsize=(16, 8))
    plt.title(f'–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–∫–∞–∑–æ–≤ –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º –≤ –¥–∏–Ω–∞–º–∏–∫–µ –¥–ª—è –∫–ª–∞—Å—Ç–µ—Ä–∞ "{cl}"')
    plt.xlabel('–ú–µ—Å—è—Ü')
    plt.xticks(rotation=0, fontsize=12)
    plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–∫–∞–∑–æ–≤')
    plt.legend(title='–ö–∞—Ç–µ–≥–æ—Ä–∏—è')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()  
    plt.show()
````

````python
# –ø–æ–¥–≥–æ—Ç–æ–≤–∏–º –≤—ã–±–æ—Ä–∫–∏ –ø–æ —Å–µ–≥–º–µ–Ω—Ç–∞–º
cluster0 = customers[customers['cluster_km'] == 0]
cluster1 = customers[customers['cluster_km'] == 1]
cluster2 = customers[customers['cluster_km'] == 2]
````

````python
# –ø—Ä–æ–≤–µ—Ä–∏–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
cluster0['avg_revenue'].hist(bins=50, alpha=0.8)
plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤—ã—Ä—É—á–∫–∏ –≤ cluster0')
plt.xlabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ')
plt.ylabel('–í—ã—Ä—É—á–∫–∞')
plt.show()

cluster1['avg_revenue'].hist(bins=50, alpha=0.8)
plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤—ã—Ä—É—á–∫–∏ –≤ cluster1')
plt.xlabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ')
plt.ylabel('–í—ã—Ä—É—á–∫–∞')
plt.show()

cluster2['avg_revenue'].hist(bins=50, alpha=0.8)
plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤—ã—Ä—É—á–∫–∏ –≤ cluster2')
plt.xlabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ')
plt.ylabel('–í—ã—Ä—É—á–∫–∞')
plt.show()
````

–¢–∞–∫ –∫–∞–∫ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å–∫–æ—Ä–µ–µ —ç–∫—Å–ø–æ–Ω–µ–Ω—Ü–∏–∞–ª—å–Ω—ã–µ, –≤–æ—Å–ø–æ–ª—å–∑—É–µ–º—Å—è –º–µ—Ç–æ–¥–æ–º –ú–∞–Ω–Ω–∞-–£–∏—Ç–Ω–∏ –¥–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è –¥–≤—É—Ö –≤—ã–±–æ—Ä–æ–∫.

–ù–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é —Å —Ü–∏–∫–ª–∞–º–∏, –ø—Ä–æ–≤–µ—Ä—è—é—â–∏–µ –≤—Å–µ 3 –∫–ª–∞—Å—Ç–µ—Ä–∞ –¥—Ä—É–≥ —Å –¥—Ä—É–≥–æ–º

<div class="alert alert-success">
  
  <b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ç–∏–º–ª–∏–¥–∞</b>
    
  ‚úîÔ∏è –í–µ—Ä–Ω–æ! –ö—Ä–∏—Ç–µ—Ä–∏–π –ú–∞–Ω–Ω–∞-–£–∏—Ç–Ω–∏ –Ω–∞–º —Ç—É—Ç –ø–æ–¥–æ–π–¥–µ—Ç!
 
</div>

````python
def hypothesis(dataframes, alpha):
    n = len(dataframes)

    for i in range(n):
        for j in range(i + 1, n):
            df_1 = dataframes[i]
            df_2 = dataframes[j]
            results = st.mannwhitneyu(df_1, df_2)
            print(f'–°—Ä–∞–≤–Ω–µ–Ω–∏–µ –≤—ã–±–æ—Ä–æ–∫ {i + 1} –∏ {j + 1}:')
            print('p-—É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏:', alpha)
            print('p-–∑–Ω–∞—á–µ–Ω–∏–µ:', results.pvalue)

            if results.pvalue < alpha:
                print('–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É\n')
            else:
                print('–ù–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É\n')
````

–ü—Ä–∏ —É—Å–ª–æ–≤–∏–∏ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –º–∞–ª–µ–Ω—å–∫–∏—Ö –≤—ã–±–æ—Ä–æ–∫ –∫–∞–∫ –≤ –Ω–∞—à–µ–º —Å–ª—É—á–∞–µ, —Å—Ç–æ–∏—Ç –ø—Ä–∏–º–µ–Ω–∏—Ç—å alpha = 0.01 –∏ —Ä–∞—Å—Å—á–∏—Ç–∞—Ç—å –¥–ª—è –Ω–µ–µ —Å–∫–æ—Ä—Ä–µ–∫—Ç–∏—Ä–æ–≤–∞–Ω–Ω—É—é –≤–µ—Ä—Å–∏—é –º–µ—Ç–æ–¥–æ–º –ë–æ–Ω—Ñ–µ—Ä—Ä–æ–Ω–∏ –¥–ª—è 3 —Ç–µ—Å—Ç–æ–≤:

````python
alpha = 0.01
m = 3
alpha_corr = alpha / m
alpha = alpha_corr

print(f'–°–∫–æ—Ä—Ä–µ–∫—Ç–∏—Ä–æ–≤–∞–Ω–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∞—è –∑–Ω–∞—á–∏–º–æ—Å—Ç—å —Ä–∞–≤–Ω–∞ {alpha}')
````

````python
hypothesis([cluster0['avg_revenue'], cluster1['avg_revenue'], cluster2['avg_revenue']], alpha)
````
